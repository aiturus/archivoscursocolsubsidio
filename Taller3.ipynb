Dando continuidad con la primera entrega del proyecto, en esta oportunidad el estudiante debe realizar las siguientes validaciones utilizando la sentencia condicional IF.

* Si el empleado es mayor de 55 años disfrutará de un bono de prepensión correspondiente al 5% de su sueldo básico.

* Si el empleado es casado y tiene hijos se le otorgará un paseo cada diciembre

* Si el sueldo básico está entre 1000000 y 1500000 tendrá una comisión del 2% sobre el valor del sueldo; Si el sueldo básico está entre 1500001 y 2000000 tendrá una comisión del 5% sobre el valor del sueldo; para todos los demás casos no habrá comisión.

* Si el empleado trabajó más de 20 días al mes y su sueldo es menor a 1000000 tendrá derecho a un bono de alimentación.

```python

id = input("Ingrese su numero de identificacion: ")
name = input("Ingrese sus nombres: ")
lastName = input("Ingrese sus apellidos: ")
address = input("Ingrese su direccion: ")
telephone = input("Ingrese su numero de telefono: ")
age = int(input("Ingrese su edad: "))
maritalStatus = input("Indique su estado civil: ")
children = int(input("Numero de hijos: "))
stature = input("Indique su estatura en centímetros: ")
jobDate = input("Indique su fecha de contratación (Día/mes/año): ")
salary = float(input("Sueldo básico: "))
daysWorked = int(input("Días Laborados: "))

prepension = salary * 0.05
comision1 = salary * 0.02
comision2 = salary * 0.05


if age > 55:
  confPrepension = "Si, por un valor de ${:,.0f}".format(prepension)
else:
  confPrepension = "No aplica"

if maritalStatus == "Casado" and children > 0:
  paseo = "Si aplica"
else:
  paseo = "No aplica"

if salary >= 1000000 and salary <= 1500000:
  comSalary = "${:,.0f}".format(comision1)
elif salary > 1500000 and salary <= 2000000:
  comSalary = "${:,.0f}".format(comision2)
else:
  comSalary = "No aplica"

if daysWorked > 20 and salary < 1000000:
  bonoAlimentacion = "Si aplica"
else:
  bonoAlimentacion = "No aplica"


print("\n Valide la informacion ingresada \n\n",
      "No de identificacion :", id, "\n", 
      "Nombres:         ", name, "\n",
      "Apellidos:       ", lastName, "\n",
      "Dirección:       ", address, "\n",
      "Teléfono:        ", telephone, "\n",
      "Edad:            ", age, "\n",
      "Estado Civil:    ", maritalStatus, "\n",
      "No de hijos:     ", children, "\n",
      "Estatura en centimetros:             ", stature, "\n",
      "Fecha de contratación (Día/mes/año): ", jobDate, "\n",
      "Sueldo básico:   ${:,.0f}".format(salary), "\n",
      "Días Laborados: ", daysWorked, "\n\n",
      "Bono Prepension:         ", confPrepension, "\n",
      "Paseo en Diciembre:      ", paseo, "\n",
      "Comision por salario:    ", comSalary, "\n",
      "Bono de alimentación:    ", bonoAlimentacion, "\n")

```
